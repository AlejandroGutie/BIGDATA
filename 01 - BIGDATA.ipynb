#**Introducción**

**Qué es Big Data?**

"Conjunto de **técnicas** que permiten **analizar, procesar y gestionar** conjuntos de datos **extremadamente grandes** que pueden ser analizados informáticamente para **revelar patrones, tendencias y asociaciones**, especialmente en relación con la conducta humana y las interacciones con los usuarios"

Tomado de: https://dej.rae.es/lema/big-data

**Las Vs del Big Data**

Las características que debe tener un proyecto para que sea considerado como Big Data se definen en función de palabras que empiezan por la letra V.
Las Vs principales del Big Data son:
* Volumen
* Velocidad
* Variedad

#**Arquitecturas de Big Data**

Los nuevos desafíos que trae consigo Big Data han generado evolución en diversas áreas, entre ellas, los sismas de información.

Veníamos acostumbrados a trabajar en sistemas centralizados en los cuales tanto almacenamiento como procesamiento dependían de una sola máquina.

Big Data propone una arquitectura diferente en la que pasamos de sistemas centralizados a sistemas de almacenamiento y procesamiento distribuidos

```
Image("local/imgs/arquitectura.png")
```

Ese cambio de arquitectura trae consigo la necesidad de que los sistemas sean escalables, y esa escalabilidad se realiza de manera horizontal de modo que si se requiere mayor procesamiento o almacenamiento basta con adicionar más máquinas.
